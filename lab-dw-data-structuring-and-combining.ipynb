In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd

url_1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url_2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url_3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df_1 = pd.read_csv(url_1)
df_2 = pd.read_csv(url_2)
df_3 = pd.read_csv(url_3)
```

Primero se estandarizan los nombres de las columnas por separado

```python
df_1.columns = df_1.columns.str.lower().str.replace(" ", "_")
df_1 = df_1.rename(columns = {"st" : "state"})
df_1.head()
```

```python
df_2.columns = df_2.columns.str.lower().str.replace(" ", "_")
df_2 = df_2.rename(columns = {"st" : "state"})
df_2.head()
```

```python
df_3.columns = df_3.columns.str.lower().str.replace(" ", "_")
df_3 = df_3.rename(columns = {"st" : "state"})
df_3.head()
```

Se combinan los tres DataFrames

```python
df = pd.concat([df_1, df_2, df_3], axis=0)
df
```

Se realizará la limpieza de datos
Primero se eliminarán filas con todos los valores nulos, basandose en la columna "customer"

```python
df = df.dropna(subset = ["customer"])
```

Se sustiturán ciertos valores por otros estandarizados

```python
df["gender"] = df["gender"].str.strip().str.upper().replace({"FEMALE": "F", "MALE": "M", "FEMAL": "F"})
```

```python
state_mapping = {"AZ": "Arizona", "Cali": "California", "WA": "Washington"}
df["state"] = df["state"].replace(state_mapping)
```

```python
df["education"] = df["education"].replace({"Bachelors": "Bachelor"})
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].str.replace("%", "").astype(float)
```

```python
df["vehicle_class"] = df["vehicle_class"].replace({"Sports Car": "Luxury", "Luxury SUV": "Luxury", "Luxury Car": "Luxury"})
```

```python
df.sample(20)
```

Se sustituirán los valores nulos de diversas columnas

```python
#Se sustituirán los valores nulos de la columna "mcustomer_lifetime_value" por la media de esa misma columna
mean_customer_lifetime_value = df["customer_lifetime_value"].mean()
df["customer_lifetime_value"].fillna(mean_customer_lifetime_value, inplace = True)
```

```python
#Con la columna "income" se realizará la misma acción
mean_income = df["income"].mean()
df["income"].fillna(mean_income, inplace = True)
```

```python
#Los valores nulos de la columna "gender" se sustituirán por "U"
df["gender"] = df["gender"].fillna("U")
```

```python
df.sample(20)
```

Se corregirán los tipos de diversas columnas

```python
df.dtypes
```

```python
#La columna number_of_open_complaints
def extract_middle_value(val):
    if pd.isna(val):
        return 0
    if isinstance(val, str) and "/" in val:
        return int(val.split("/")[1])
    try:
        return int(val)
    except(ValueError, TypeError):
        return 0

df["number_of_open_complaints"] = df["number_of_open_complaints"].apply(extract_middle_value)
df["number_of_open_complaints"] = pd.to_numeric(df["number_of_open_complaints"])
```

```python
#Las columnas, "income", "monthy_premium_auto" y "total_claim_amount" se pasarán de float a int
float_columns = ["income", "monthly_premium_auto", "total_claim_amount"]

for column in float_columns:
    df[column] = df[column].fillna(0).astype("int64")
```

```python
df.sample(20)
```

```python
df.duplicated()
```
